```python
#Import dependencies
import pandas as pd

#Import modules
from sqlalchemy import create_engine

```

```python
#Load CSV files
clean_inspection = "Resources/clean_inspection_df2.csv"
clean_income = "Resources/clean_income_df.csv"
```

```python
#View inspection dataframe
clean_inspection_df = pd.read_csv(clean_inspection)
clean_inspection_df.head()
```

```python
#Rename columns
clean_inspection_df.rename(columns = {'CUISINE DESCRIPTION':'CUISINE_DESCRIPTION', 'GRADE DATE': 'GRADE_DATE'}, inplace = True)
```

```python
#Rearrange inspection dataframe column
new_inspection_df = clean_inspection_df[["Borough", "DBA", "STREET", "CUISINE_DESCRIPTION", "SCORE", "GRADE", "GRADE_DATE"]]
new_inspection_df
```

```python
#View income dataframe
clean_income_df = pd.read_csv(clean_income)
clean_income_df.head()
```

```python
#Rename columns
clean_income_df.rename(columns = {'INCOME LEVEL':'INCOME_LEVEL', 'GRADE DATE': 'GRADE_DATE'}, inplace = True)
```

```python
#Rearrange income dataframe column
new_income_df = clean_income_df[["Borough", "DBA", "INCOME_LEVEL", "STREET",  "GRADE_DATE", "LATITUDE", "LONGITUDE"]]
new_income_df
```
